```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import data
from otsu import otsu
%load_ext autoreload
%autoreload 2
```

```python
img = data.camera()

# Otsu's thresholding
ret2,th2 = cv2.threshold(img,30,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# plot all the images and their histograms
images = [img, 0, th2]
titles = ['Original Noisy Image','Histogram',"Otsu's Thresholding"]


plt.subplot(3,3,1),plt.imshow(images[0],'gray')
plt.title(titles[0])
plt.subplot(3,3,2),plt.hist(images[0].ravel(),256)
plt.title(titles[1])
plt.subplot(3,3,3),plt.imshow(images[2],'gray')

plt.title(titles[2]), plt.xticks([]), plt.yticks([])
plt.gcf().set_size_inches((15,12))
plt.show()
```
